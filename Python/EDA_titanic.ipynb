```python
from sklearn.model_selection import train_test_split
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
import pandas as pd

# Splitting the data into train and test sets
x_train, x_test, y_train, y_test = train_test_split(
    df.drop(columns=["Survived"]),
    df["Survived"],
    train_size=0.8,
    random_state=2
)

# Creating a transformer to handle missing values and one-hot encoding
transformer = ColumnTransformer(transformers=[
    ("Imputation", SimpleImputer(strategy="constant", fill_value=28),["Age"]),
    ("Ohe", OneHotEncoder(drop="first", sparse_output=False), ["Sex"])
], remainder="passthrough")

x_train_ =transformer.fit_transform(x_train)

x_test_ =transformer.transform(x_test)




# Convert x_train_ and x_test_ arrays to pandas DataFrames
#Transformer object has attribute get_feature_names_out(input_features=x_train.columns)
# remember here input_features comes from x_train before fitting into the transformer.
# columns (new column names) = transformer_object.get_feature_names_out(input_features=x_trian.columns)

x_train_df = pd.DataFrame(x_train_, columns=transformer.get_feature_names_out(input_features=x_train.columns))
x_test_df = pd.DataFrame(x_test_, columns=transformer.get_feature_names_out(input_features=x_train.columns))

# Display the resulting DataFrames
print("x_train_df:")
print(x_train_df.head())

print("\nx_test_df:")
print(x_test_df.head())
```
