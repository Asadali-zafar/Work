**Hypothesis Testing**

```python
import scipy.stats as stats

fraudulent_amounts = data[data['isFraud'] == 1]['amount']
non_fraudulent_amounts = data[data['isFraud'] == 0]['amount']

t_stat, p_value = stats.ttest_ind(fraudulent_amounts, non_fraudulent_amounts, equal_var=False)

alpha = 0.05


if p_value < alpha:
    print("There is a significant difference in the mean 'amount' between fraudulent and non-fraudulent transactions.")
    print(f"p-value: {p_value:.6f}")
else:
    print("There is no significant difference in the mean 'amount' between fraudulent and non-fraudulent transactions.")
```

```python

fraudulent_amounts = data[data['isFraud'] == 1]['step']
non_fraudulent_amounts = data[data['isFraud'] == 0]['step']


t_stat, p_value = stats.ttest_ind(fraudulent_amounts, non_fraudulent_amounts, equal_var=False)

alpha = 0.05

if p_value < alpha:
    print("There is a significant difference in the mean 'step' between fraudulent and non-fraudulent transactions.")
    print(f"p-value: {p_value:.6f}")
else:
    print("There is no significant difference in the mean 'step' between fraudulent and non-fraudulent transactions.")
```

**Model application**

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
```

```python
y = numerical_data['isFraud']
X = numerical_data.drop(columns=['isFraud'])
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, random_state=42)
```

```python
X_train
```

```python
# Initialize the Random Forest Classifier
rf_classifier = RandomForestClassifier(n_estimators=100, random_state=42)

# Fit the classifier on the training data
rf_classifier.fit(X_train, y_train)
```

```python
y_pred = rf_classifier.predict(X_test)

accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy: {accuracy}")

print(classification_report(y_test, y_pred))

confusion = confusion_matrix(y_test, y_pred)
print("Confusion Matrix:\n", confusion)
```
